```python
import requests 
import time
from bs4 import BeautifulSoup 
```

```python
url = 'https://en.wikipedia.org/wiki/Lionel_Messi' 
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser') 
```

```python
list = []
for i in soup.select('p'):
    list.append(i.getText())
#     print(i.getText())
print(list)
```

```python
sentences = []
for i in list:
    l = i.split('.')
    for j in l:
        sentences.append(j)
sentences
```

```python
for i in sentences:
    count = 0
    for j in i:
        if(j==" "):
            count = count+1
    if(count<3):
        sentences.remove(i)
sentences
```
